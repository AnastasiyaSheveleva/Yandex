При изучении данных в таблице видим, что данные стажа во многих строках имеют отрицательные значения. Вероятно такаяаномалия возникла из-за объединения данных из разных баз, в одной учет может быть в часах в другой в днях. До выполнения заполнения пропусков по столбцу "days_employed" считаю перевести данные в один формат.

### Шаг 2.1 Заполнение пропусков

```python
print(len(data[data['days_employed'].isna()])) 
print(len(data[data['total_income'].isna()]))
```

```python
days_employed_avg = data['days_employed'].median()
data['days_employed'] = data['days_employed'].fillna(value=days_employed_avg)
total_income_avg = data['total_income'].median()
data['total_income'] = data['total_income'].fillna(value=total_income_avg)
data.info()
display(data.head(15))
```

При анализе полученных данных выявлены следующие ошибки:
1. обнаружены пропуски в столбцах "общий трудовой стаж в днях" и "ежемесячный доход"
2. доля пропущенных значений составляет 10% от общего количесва данных 
[(21525-19351)/21525*100=10] в каждом столбце
3. при подаче заявки на кредит потенциальный заемщик врядли допускает пропуски в предложенной к заполнинию анкете банка. более вероятные причины пояления пропусков в файле:  - ошибка сотрудника банка при занесении данных в базу
- техническая ошибка при подготовке выгрузки данных
4. при удалении строк содержащих пропуски приведет к потере 10% данных, что может вызвать искажение информации и выводах сделанных на ее основании, поэтому лучшим выходом из сложившейся ситации считаю замену пропущенных данных медианным значением по столбцу, т.к. данные в столбцах являются колличественными переменными
5. для замены пропусков в данных выбран метод медианных значений по причине того, что среднее значение некорректно будет характеризовать данные, когда некоторые значения сильно выделяются среди большинства.

### Шаг 2.6. Категоризация дохода.

```python
def income_category(total_income):
    if total_income <= 30000:
        return 'E'
    if total_income <=50000:
        return 'D'
    if total_income <=200000:
        return 'C'
    if total_income <=1000000:
        return 'B'
    return 'A'
```

```python
data['total_income_category'] = data['total_income'].apply(income_category)
print(data['total_income_category'].value_counts())
```

### Шаг 2.7. Категоризация целей кредита.

```python
print(data['purpose'].value_counts())
```

```python
def new_purpose(cell):
    purpose_dictionary = {'операции с недвижимостью': ['недвижимостью', 'недвижимости', 'жилья', 'жильем', 'недвижимость', 'жилье'],
                         'операции с автомобилем': ['автомобиля', 'автомобиль', 'автомобилем', 'автомобили'],
                         'проведение свадьбы': ['свадьба', 'свадьбы', 'свадьбу'],
                         'получение образования':['образованием', 'образование', 'образования'],
                         'ремонт':['ремонт']}
    for category, synonym in purpose_dictionary.items():
        for syn in synonym:
            if syn in cell: 
                return category
    return None
data['purpose_category'] = data['purpose'].apply(new_purpose)
data[['purpose', 'purpose_category']]
```

```python
print(data['purpose_category'].value_counts())
```

### Ответы на вопросы.

##### Вопрос 1:

```python
report = data.pivot_table(index='children', columns='debt', values = 'total_income', aggfunc = 'count').fillna(0)
report.columns = ['Нет задолженности', 'Задолженность']
report['%'] = report['Задолженность'] / (report['Задолженность'] + report['Нет задолженности'])
report = report.sort_values(by=['%'], ascending=False)
display(report.style.format({'%': '{:.2%}', 'Задолженность': '{:.0f}', 'Нет задолженности': '{:.0f}'}))
```

##### Вывод 1:

Изходя из полученных данных - риски по невозврату кредита повышаются с появлением детей в с6емьях заемщиков. Но при этом можно заметить тенденцию - чем больше детей у заемщика, тем реже он обращается за получением кредита
